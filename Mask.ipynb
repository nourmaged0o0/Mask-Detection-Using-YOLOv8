```python
# MAIN
import cv2
from ultralytics import YOLO
import time

model = YOLO("best.pt")  # Load pre-trained model
cap = cv2.VideoCapture("path_to_video.mp4") 

# Dictionary to keep track of unique IDs and their timestamps
unique_ids = {}

# Duration to display the box in seconds
display_duration = 2

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO model on the frame
    results = model.track(frame, persist=True, conf=0.6)

    # Process each result from the model
    for result in results:
        # Loop over the detected objects
        for box in result.boxes:
            if (box.id is not None) and box.cls == 1:  # Check if ID exists
                id = int(box.id.item())  # Convert to standard Python integer
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)  # Get box coordinates

                # Check if it's a new ID
                if id not in unique_ids:
                    unique_ids[id] = time.time()
                else:
                    if time.time() - unique_ids[id] >= 1 and box.conf >= 0.50:
                        cropped_frame = frame[y1:y2, x1:x2]
                        cv2.putText(cropped_frame, f"ID: {id}", (0, 7), cv2.FONT_HERSHEY_SIMPLEX, 0.35, (0, 0, 255),thickness=1)
                        cv2.imwrite(f"C:\\Python projects\\runs\\detect\\train15\\screens\\{id}.png", cropped_frame)
                # Draw the box for new IDs if within display duration
                if time.time() - unique_ids[id] < display_duration:
                    cv2.rectangle(frame, (x1 + 4, y1 + 4), (x2 - 4, y2 - 4), (0, 255, 0), 3)
                    cv2.circle(frame, (int((x1 + x2) / 2), int((y1 + y2) / 2)), 4, (0, 255, 0), -1)
                    cv2.putText(frame, f"ID: {id} Added", (544, 37), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 2)

    # Draw the annotations on the frame
    annotated_frame = results[0].plot()

    # Display the frame with detection and crosshair
    cv2.imshow("YOLOv8 with Crosshair", annotated_frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Print all unique IDs detected during the session
print(f"Unique IDs detected: {list(unique_ids.keys())}")

cap.release()
cv2.destroyAllWindows()
```
